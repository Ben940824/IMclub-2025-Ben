### Deep Copy 和 Shallow Copy
#### 小試身手 6
用 deep copy 和 shallow copy 兩種方法分別將 a 串列複製給 b 串列，並印出結果觀察兩者的差異

```python
# Deep Copy
a = [1, 2, 3]
b = a[:]

a.pop(1)
print(a)
print(b)
```

```python
# Shallow Copy
a = [1, 2, 3]
b = a

a.pop(1)
print(a)
print(b)
```
